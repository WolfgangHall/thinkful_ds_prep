```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
plt.plot([0, 1, 2, 3])
plt.show()
```

```python
np.random.seed(1221)  # makes the random function consistent and repicable

df = pd.DataFrame()
df['rand'] = np.random.rand(100)   # add a column of random number between 0 and 1
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

# changes index
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)

plt.plot(df['rand'], color = 'purple')
plt.plot(df['rand_shift'], color = 'green')

plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')

plt.show()
```

# Scatter

```python
plt.scatter(x = df['rand'], y = df['rand_sq'], color = 'purple', marker = 'x', s = 10)
plt.scatter(x = df['rand'], y = df['rand_shift'], color = 'green', marker = 'x', s = 10)
plt.show()
```

# Pandas Plot

```python
df.plot(kind = 'scatter', x = 'counts_sq', y = 'counts_sqrt')
df.plot(kind = 'line')
```

# Subplots

```python
# first two parameters define dimensions of the plot
# third identifies the subplot you're creating
```

```python
plt.figure(figsize=(5,10))

plt.subplot(2, 1, 1)
plt.plot(df['rand'], color = 'purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(2, 1, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')

plt.tight_layout()
plt.show()
```

```python
# plt.subplot() -> (2, 1, 2)
# (2, 1)  ->  2 high, 1 wide
# third parameter defineds which plot to create
# only works for subplots with less than 10 figures

# use plt.tight_layout() for additional spacing around subplots
```
